```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# importing data
from google.colab import files
import pandas as pd

# Upload file
uploaded = files.upload()

# baca dengan pandas
data = pd.read_csv('ilustrasikm.csv')

# Lihat isi data
data.head()
```

```python
# Ambil dua kolom numerik untuk clustering
X = data[['X1', 'X2']]
```

```python
#K-Means
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

# Ambil hasil label dan centroid
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
# Plot hasil cluster
plt.figure(figsize=(8,6))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=50, label='Centroids')
plt.title('K-Means Clustering (3 Clusters)')
plt.xlabel(X.columns[0])
plt.ylabel(X.columns[1])
plt.legend()
plt.show()


```
