## Libraries
python menjadi tools yang popular untuk pengolahan data dan pengembangan model matematika dan machine learning karena tersedia banyak sekali library sebagai tambahan (add-on) yang membantu, contohnya:
- NumPy : library ini menyediakan beberapa fungsi basic linear algebra, advance number randomizer, transformasi matrix, dll
- SciPy : library ini merupakan add-on dari NumPy dengan beberapa high-level engineering and science modules, seperti sparce matrix
- Matplotlib : library data visualisasi standar pada python dengan banyak pilihan graph dan chart
- Pandas : library yang popular dalam pengolahan data, digunakan dalam proses manipulasi/transformasi data dan fase preprocessing data
- Scikit Learn: library yang memiliki banyak tools berguna untuk statistical modeling and machine learning seperti dimensionality reduction, clustering, regression, and classification.
- Seaborn: library ini membantu untuk membuat visualisasi statistik yang informatif dan menarik.
- Requests: library yang digunakan untuk mengakses web.

#### Row selection/filtering
pemilihan row atau filtering dilakukan dengan menggunakan conditional statement

#### conditional statement pada pandas

```python
mask = data_latihan['item_price'] <= 10000000
mask
```

```python
data_latihan[mask]
```

```python
data_latihan[~mask]
```
