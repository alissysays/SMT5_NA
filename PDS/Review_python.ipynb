### 8. Tipe Data Sequence

terdapat beberapa tipe data sequence, pada kali ini akan di perkenalkan 3 tipe yaitu str, list dan tuple.

#### 8.1 list
Python dapat mengelompokkan beberap tipe data yang berbeda menjadi satu kelompok, yang kemudian dikenal sebagai list. Analogi list adalah sebuah lemari dengan laci yang berurutan dan setiap laci di beri nomor mulai dari 0

#### membuat list

```python
#definition
contoh_list = [1,2,'text',4.0,68,3,5]
print(contoh_list)
```

#### mengakses data

```python
#indexing
ambil = contoh_list[3] # mengambil data ke 4 (index 3, perhitungan mulai dari 0)
print(ambil)

ambil = contoh_list[1:4] # mengambil data ke 2 sd 4 (index 1,2,3) angka setelah ':' di exclude kan
print(ambil)

ambil = contoh_list[3:] # mengambil data ke 4 sd akhir (index 3,4,5,6)
print(ambil)

ambil = contoh_list[:5] # mengambil data awal sd ke 5 (index 0,1,2,3,4)
print(ambil)

ambil = contoh_list[-2] # mengambil data ke 2 dari kanan
print(ambil)
```

List juga dapat dipisah-pisah dan dapat digabungkan, ditambahkan dan lainnya

```python
# ditambahkan
contoh_list + [....]
# penggandaan
3*contoh_list[....]
```

struktur dari list yang sudah dibuat juga bisa diubah seperti untuk mengganti elemen, menghilangkan elemen dll

```python
# mengganti nilai
contoh_list[0:1]=[2,3]
# mengetahui jumlah elemen
len()
# menambahkan anggota
___.append()
# mengurutkan
sort()
# menghapus elemen
del
dll
```

### 9. Dictionary
Analogi dictionary adalah sebuah lemari yang memiliki beberapa laci yang memiliki nama unique. Tidak seperti data sekuensial yang diurutkan menurut urutan indeks, dictionary di indeks dengan keys dari dictionary itu sendiri, dimana suatu key dapat berupa angka-angka, bilangan pecahan atau string.

#### 9.1 membuat dict
untuk membuat dictionary hanya memerlukan sepasang tanda kurung kurawal **{ }**

```python
contoh_dict = {'Dale': 9073, 'Page': 9128, 'Small': 9223, 'Nate':9330}
print(contoh_dict)
```

#### 9.2 mengakses data

```python
ambil = contoh_dict['Page']
print(ambil)

# mengambil semua data
print(contoh_dict.values())

# mengambil semua keys
print(contoh_dict.keys())
```
